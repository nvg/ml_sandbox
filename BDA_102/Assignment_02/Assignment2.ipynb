```python
import os
import pandas as pd
import nltk
```

```python
df=pd.read_csv('jokes_2.csv')
data=df.drop(['Question','ID'], axis=1)
```

```python
data.head(n=10)
```

# Data Cleaning

```python
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
```

```python
i=0
for txt in data['Answer']:
     w_list = re.sub(r'[^a-z\s]',' ',txt.lower()).split()
     w_list=[w for w in w_list if w not in stopwords.words('english')]
     data.iloc[i]['Answer']=' '.join(w_list)
     i+=1
```

```python
data.head()
```

# TF-IDF

```python
# Term Frequency Inverse Document Frequency
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import numpy as np
```

```python
vectorizer = TfidfVectorizer(sublinear_tf=True,min_df=1, stop_words='english')
term_doc_matrix = vectorizer.fit_transform(data['Answer'])
feature_names = vectorizer.get_feature_names()
```

```python
term_doc_matrix
```

```python
feature_names
```

```python
# np.shape(term_doc_matrix)
dense_term_doc_matrix = term_doc_matrix.todense()

sum_tdm = dense_term_doc_matrix.sum(axis=0)

import numpy as np
sort_list=np.argsort(sum_tdm)
for i in range(0,10,1):
    print(feature_names[np.array(sort_list)[0,i]])
```

# LSA - SVD

```python
# Latent Semantic Analysis - LSA takes TFIDF one step further by use SVD to perform dimentionality reduction
# Singular Value Decomposition

from scipy.sparse.linalg import svds
u,s,vt=svds(term_doc_matrix)
```

```python
s.shape
```

```python
rank=2
uk=u[:,0:rank]
sk=np.diag(s)[0:rank,0:rank]
vk=vt[0:rank,:]
```

```python
vk.shape
```

```python
len(vectorizer.get_feature_names())
```

```python
aa=np.argsort(vk,axis=1)
for t in range(rank):
    print('------ topic number' + str(t) + '--------')
    for i in range(10):
        print(vectorizer.get_feature_names()[aa[t,i]])
```
