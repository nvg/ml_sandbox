```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
```

For this example, use a linear activation function within the keras library to create a regression-based neural network. 
Essentially, we are trying to predict the value of a potential car sale (i.e. how much a particular person will spend on buying a car) for a customer based on the following attributes:

-Age

-Gender

-Average miles driven per day

-Personal debt

-Monthly income

```python
dataset=np.loadtxt("cars.csv", delimiter=",")
x=dataset[:,0:5]
y=dataset[:,5]
y=np.reshape(y, (-1,1))
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
print(scaler_x.fit(x))
print(scaler_y.fit(y))
xscale=scaler_x.transform(x)
yscale=scaler_y.transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(xscale, yscale)
```

# Create the model.

```python
model = Sequential()
model.add(Dense(units=12, input_dim=5, kernel_initializer='normal', activation='relu'))
model.add(Dense(units=8, activation='relu'))
model.add(Dense(units=1, activation='linear'))
```

```python
model.summary()
```

# compile/fit the model

```python
model.compile(loss='mse', optimizer='adam', metrics=['mse', 'mae'])

# model.fit(x, y, epochs=100,batch_size=10)
model.fit(X_train, y_train, epochs=1000, batch_size=100,  verbose=1, validation_split=0.2)
```

# predict the new set of data

```python
Xnew = np.array([[40, 0, 26, 9000, 8000]])
```

```python
ypred = model.predict(scaler_x.transform(Xnew))
ypred = scaler_y.inverse_transform(ypred)

print("%s = %s" % (Xnew, ypred))
```
