```python
df.drop(['ID', 'Source', 'TMC', 'End_Lat', 'End_Lng', 'Number', 'Timezone', 'Description', 'Airport_Code'], axis=1, inplace = True)
```

```python
df.info()
```

```python
for group in ['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)']:
    df[group] = df[group].fillna(df.groupby('State')[group].transform('mean'))
```

```python
for group in ['City', 'Zipcode', 'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight', 'Weather_Timestamp']:
    df = df[df[group].notna()]
```

```python
for group in ['Wind_Direction', 'Weather_Condition']:
    df[group].fillna('Unavailable', inplace=True)
```

```python
df['Precipitation(in)'].fillna(0, inplace=True)
```
