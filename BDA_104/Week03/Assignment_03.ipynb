```python
# Part I: 
# 1.	Import csv dataset (MOVIE DATASET) 
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
import json
def get_data(column, df, key):
    df_temp = pd.DataFrame([])
    for i in column:
        json_data = json.loads(i)
        df1 = pd.DataFrame.from_dict(json_data)
        if len(i) > 2:
            df2 = pd.DataFrame([df1[key][0]])
        else:
            df2 = df1
        df_temp = pd.concat([df_temp, df2], ignore_index=True)
    df_temp.columns = [column.name + '_' + key]

    df = pd.concat([df, df_temp], axis=1)
    return df
```

```python
df.head()
```

```python
for i in ['genres', 'spoken_languages', 'production_companies', 'production_countries']:
    df = get_data(df[i], df, 'name')
```

```python
# 2.	Identify what features have missing values 
df.head()
```

```python
na_df = df[df.isnull().any(axis=1)]
na_df.info()
na_df.head(100)
```

```python
na_df = na_df.drop(['homepage', 'tagline'], axis=1)
na_df = na_df[na_df.isnull().any(axis=1)]
na_df.info()
na_df.head(100)
```

```python
na_df.info()
na_df.head(100)
```

```python
# Features that are missing values are 'homepage', 'tagline', 'runtime', 'release_date', 'overview', where 
# homepage and tagline represent the largest set of missing values
```

```python
# 3.	Replace missing values with 0
df = df.fillna(0)
```

```python
# 4.	Split dataset into training and test datasets in 0.7/0.3
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df.drop('vote_average', 1)
y = df['vote_average']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# 5.	Select all numerical features that can be used to predict ‘vote average’ target 
# numerical features would be budget and popularity
X.info()
```

```python
# 6.	Add three categorical features (for example, original language, status and genres) and transform them into numerical 
for i in ['genres_name', 'spoken_languages_name', 'production_companies_name', 'production_countries_name']:
    X[i] = X[i].astype('category')
    X[i].head()
    X[i] = X[i].cat.codes

X.info()
```

```python
# 7.	Fill the training and test data with required information: 
# a.	Use: train[features].dropna() 
X = X.dropna()
X.drop(['id', 'genres', 'homepage', 'keywords', 'original_language', 'original_title', 'overview', 'production_companies', 'production_countries', 'release_date', 'spoken_languages', 'status', 'tagline', 'title'], axis=1, inplace=True)
X.info()
```

```python
# 8.	Import linear regression function from sklearn: 
# a.	Use: linear_model.LinearRegression() from sklearn
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

# redo the split with the updated data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
reg.fit(X_train, y_train)

print(' The estimated intercept coefficient is %.2f ' %reg.intercept_)
print(' The number of coefficients used was %d ' % len(reg.coef_))
```

```python
y_pred = reg.predict(X_test)

from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R^2', metrics.r2_score(y_test, y_pred))
```

```python
# 9.	Use score to identify most accurate and predictive features. Justify the final choice of the features.
coeff_df = pd.DataFrame(reg.coef_, X.columns, columns=['Coefficient'])  
coeff_df.sort_values(by=['Coefficient'])
coeff_df
```

```python
# Categories affecting the score are: budget (negative corr) and popularity (positive corr)
```

```python
# Step 10

from sklearn.feature_selection import VarianceThreshold
Xvt = X.copy(deep=True)
# X = [[0, 0, 1], [0, 1, 0], [1, 0, 0], [0, 1, 1], [0, 1, 0], [0, 1, 1]] 
sel = VarianceThreshold(threshold=(.8 * (1 - .8))) 
sel.fit_transform(Xvt)
```

```python
regvt = LinearRegression()

# redo the split with the updated data
X_train, X_test, y_train, y_test = train_test_split(Xvt, y, test_size = 1/3, random_state = 0)
regvt.fit(X_train, y_train)

print(' The estimated intercept coefficient is %.2f ' %regvt.intercept_)
print(' The number of coefficients used was %d ' % len(regvt.coef_))

y_pred = regvt.predict(X_test)

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R^2', metrics.r2_score(y_test, y_pred))

coeff_df = pd.DataFrame(regvt.coef_, Xvt.columns, columns=['Coefficient'])  
coeff_df.sort_values(by=['Coefficient'])
coeff_df
```

```python
# Similar results as without VarianceTrheshold
```

```python
# 11.	Recursive feature elimination with linear regression: 
from sklearn.feature_selection import RFE
```

```python
regrfe = LinearRegression()
Xrfe = X.copy(deep=True)

# redo the split with the updated data
X_train, X_test, y_train, y_test = train_test_split(Xrfe, y, test_size = 1/3, random_state = 0)
regrfe.fit(X_train, y_train)

# Create the RFE object
rfe = RFE(estimator=regrfe, n_features_to_select=1, step=1)
rfe.fit(X_train, y_train)
```

```python
# print summaries for the selection of attributes
print(X_train.columns)
print(rfe.support_)
print(rfe.ranking_)
```
