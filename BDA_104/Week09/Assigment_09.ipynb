### 1. Define K-means clustering? 

K-means groups similar data points together to discover patterns. We initaially set a fixed number (k) of clusters in a dataset. A cluster refers to a collection of data points aggregated together because of certain similarities (e.g. euclidian distance). To choose the optimal number of clusters use the elbow method.

K-means algorithm works as follows:

  1. Randomly assign a number, from 1 to K, to each of the observations. These serve as initial cluster assignments for the observations. 
  2. Iterate until the cluster assignments stop changing:
    * For each of the K clusters, compute the cluster centroid. 
    * Assign each observation to the cluster whose centroid is closest.   

### 2. K-means example

```python
%matplotlib inline
import random as rnd
import numpy as np

import matplotlib.pyplot as plt

from scipy import stats

from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# use seaborn plotting defaults
import seaborn as sns; 
sns.set()
```

```python
# Use a random number generator to generate the centre value between 2-10
center = rnd.randint(2, 10)
```

```python
# Use a random number generator to generate cluster_std between 0.01 and 0.99 
cluster_std = 0.01 + (rnd.random() * (0.99 - 0.01))
```

```python
# Use the make_blob function to generate a new data set
X, y = make_blobs(n_samples=300, centers=center,
                  random_state=0, cluster_std=cluster_std)

plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
# Use a loop structure (for or while) that applies K-means with 
# different number of means (k takes the value between 2 and 15)

wcss = []
sscr = []
for i in range(2, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    preds = kmeans.fit_predict(X)
    wcss.append(kmeans.inertia_)
    sscr.append(silhouette_score(X, preds))
    
plt.plot(range(2, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

plt.plot(range(2, 15), sscr)
plt.title('Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('S_S')
plt.show()
```

```python
# Choose the k-value that has the highest silhouette score
print(sscr.index(max(sscr)), max(sscr))
```

```python
# Print the following values:
# Centres and the k value with the highest silhouette score
# Repeat for 100 times
```

```python
for i in range(100):
    X, y = make_blobs(n_samples=300, centers=rnd.randint(2, 10), random_state=0, cluster_std=(0.01 + (rnd.random() * (0.99 - 0.01))))
    plt.scatter(X[:, 0], X[:, 1], s=50)
    plt.show()
    
    sscr = []
    for j in range(2, 15):
        kmeans = KMeans(n_clusters = j, init = 'k-means++', random_state = 42)
        preds = kmeans.fit_predict(X)
        sscr.append(silhouette_score(X, preds))
        
    print("Iter=%s, Best K-value= %s, Score = %s" % (i, sscr.index(max(sscr)) + 2, max(sscr)))
```
