```python
# 3. Apply following data preparation rules
# # Drop the columns where all elements are missing values:

# let's examine the NA values
na_df = df[df.isnull().any(axis=1)]
na_df.info()
```

```python
# There are 708 rows with NA values - dropping all would not be a good idea
# let's look at what's missing
na_df.head(100)
```

```python
# Mostly Cabin & Age are missing 
# this is really depends on what we are looking for
# but as a generic advice, it might be a good idea to drop Cabin altogether
# and use mean value for filling in missing age

#
# For the sake of this excersie let's attempt to drop using different options
#
dropped_any_df = df.dropna(how='any')
dropped_any_df.info()
dropped_any_df.head()
```

```python
dropped_th_2 = df.dropna(thresh=2)
dropped_th_2.info()
dropped_th_2.head()
```

```python
# # Fill all missing values with the mean of the particular column
df_mean = df.fillna(df.mean())
df_mean.info() # all rows are retrained
df_mean.head()
```

```python
# # Fill any missing value in column 'A' with the column median
df_med_age = df.copy(deep=True)
df_med_age['Age'] = df_med_age['Age'].fillna(df_med_age['Age'].median())
df_med_age.head()
```
